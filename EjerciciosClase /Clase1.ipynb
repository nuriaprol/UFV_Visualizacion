```python
import plotly_express as px
```

```python
gapminder = px.data.gapminder()
```

```python
gapminder.head()
```

```python
gapminder2007 = gapminder.query("year == 2007")
gapminder2007.head()
```

```python
fig = px.scatter(gapminder2007, x="gdpPercap", y="lifeExp", color = "continent")
fig.show()
```

```python
fig = px.scatter(gapminder2007, x="gdpPercap", y="lifeExp", color = "continent", hover_name ="country")
fig.show()
```

```python
fig = px.scatter(gapminder2007, x="gdpPercap", y="lifeExp", color = "continent", hover_name ="country", labels = {"lifeExp":"Life expectation", "gdpPercap":"GDP Per Capita (USD)"}, size = "pop", size_max = 60)
fig.show()
```

```python
fig = px.scatter(gapminder2007, x="gdpPercap", y="lifeExp", color = "continent", hover_name ="country", labels = {"lifeExp":"Life expectation", "gdpPercap":"GDP Per Capita (USD)"},
                 size = "pop", size_max = 60, facet_col="continent" )
fig.show()
```

```python
fig2 = px.choropleth(gapminder2007, locations="iso_alpha", color="lifeExp", hover_name="country", animation_frame="year", color_continuous_scale=px.colors.sequential.Plasma, projection="natural earth")
fig2.show()
```
